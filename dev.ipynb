```python
from google_play_scraper import Sort, reviews
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Function to get reviews for a given app
def get_app_reviews(app_id, count=50):
    result, continuation_token = reviews(
        app_id,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        count=count,
    )

    result, _ = reviews(
        app_id,
        continuation_token=continuation_token
    )

    return result

# Replace 'com.fantome.penguinisle' with the actual ID of your app
app_id = 'com.nekki.shadowfight'

# Get app reviews
app_reviews = get_app_reviews(app_id)

# Create a dataset with the fetched reviews
dataset = [
    {'text': review['content'], 'label': 0 if review['score'] >= 3 else 1}
    for review in app_reviews
]

# Split the dataset into training and testing sets
train_data, test_data = train_test_split(dataset, test_size=0.2, random_state=42)

# Use the same Naive Bayes pipeline as before
vectorizer = CountVectorizer()
classifier = MultinomialNB()
nb_pipeline = make_pipeline(vectorizer, classifier)

# Train the Naive Bayes model on the training data
nb_pipeline.fit([item['text'] for item in train_data], [item['label'] for item in train_data])

# Predict using the Naive Bayes model
predictions = nb_pipeline.predict([item['text'] for item in test_data])

# Calculate Naive Bayes accuracy
accuracy = accuracy_score([item['label'] for item in test_data], predictions)
print(f'Naive Bayes Accuracy: {accuracy * 100:.2f}%')

# Iterate through the reviews and print relevant information along with fake review prediction
for index, review in enumerate(app_reviews):
    prediction = nb_pipeline.predict([review['content']])[0]
    print(f"Review #{index + 1} - Prediction: {'Fake' if prediction == 1 else 'Genuine'}")
    print("User Name:", review['userName'])
    print("Content:", review['content'])
    print("Score:", review['score'])
    print("\n" + "=" * 50 + "\n")  # Separator for better readability
```

```python
from transformers import pipeline
from google_play_scraper import app, Sort, reviews

def get_app_reviews(app_id, count=50):
    result, continuation_token = reviews(
        app_id,
        lang='en',
        country='us',
        sort=Sort.MOST_RELEVANT,
        count=count,
    )

    result, _ = reviews(
        app_id,
        continuation_token=continuation_token
    )

    return result

# Load sentiment analysis pipeline
sentiment_analysis_pipeline = pipeline("sentiment-analysis", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")

app_id = 'com.nekki.shadowfight'
app_reviews = get_app_reviews(app_id)

# Initialize counters
positive_reviews = 0
neutral_reviews = 0
negative_reviews = 0

# Set sentiment score threshold
threshold = 0.5

# Separate dictionaries for positive, neutral, and negative reviews
positive_dict = {}
neutral_dict = {}
negative_dict = {}

for review in app_reviews:
    name = review['userName']
    date = review['at']
    review_body = review['content']

    try:
        # Perform sentiment analysis
        sentiment_result = sentiment_analysis_pipeline(review_body)

        if sentiment_result:
            raw_sentiment_label = sentiment_result[0].get('label', 'UNKNOWN')
            sentiment_score = sentiment_result[0].get('score', 0)
        else:
            raw_sentiment_label = 'UNKNOWN'
            sentiment_score = 0

        # Apply threshold for sentiment classification
        if sentiment_score > threshold:
            sentiment_label_text = 'POSITIVE'
            positive_reviews += 1
            positive_dict[name] = {'Date': date, 'Review': review_body, 'Raw Sentiment Label': raw_sentiment_label, 'Sentiment Score': sentiment_score}
        elif sentiment_score < (1 - threshold):
            sentiment_label_text = 'NEGATIVE'
            negative_reviews += 1
            negative_dict[name] = {'Date': date, 'Review': review_body, 'Raw Sentiment Label': raw_sentiment_label, 'Sentiment Score': sentiment_score}
        else:
            sentiment_label_text = 'NEUTRAL'
            neutral_reviews += 1
            neutral_dict[name] = {'Date': date, 'Review': review_body, 'Raw Sentiment Label': raw_sentiment_label, 'Sentiment Score': sentiment_score}

        # Print detailed information
        print(f"Name: {name}")
        print(f"Date: {date}")
        print(f"Review: {review_body}")
        print(f"Raw Sentiment Label: {raw_sentiment_label}")
        print(f"Sentiment Score: {sentiment_score:.4f}")
        print(f"Sentiment: {sentiment_label_text}")
        print("\n")

    except Exception as e:
        print(f"Error processing review: {e}")

# Print summary
print(f"Total Reviews: {len(app_reviews)}")
print(f"Positive Reviews: {positive_reviews}")
print(f"Neutral Reviews: {neutral_reviews}")
print(f"Negative Reviews: {negative_reviews}")

# Additional: Print negative reviews dictionary
print("\nNegative Reviews Dictionary:")
print(negative_dict)
```
